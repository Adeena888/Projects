# Import

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

# Load and analyse data

```python
df = pd.read_csv('weather_analysis_dataset.csv')
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.shape
```

```python
cr = df[['Temperature','Humidity','Wind Speed']].corr()
cr
```

# Find max and min

```python
temp_max = df['Temperature'].max()
temp_min = df['Temperature'].min()

humidity_max = df['Humidity'].max()
humidity_min =df['Humidity'].min()

wind_max = df['Wind Speed'].max()
wind_min = df['Wind Speed'].min()
```

```python
print(f'-Maximum temperature: {temp_max}째C -> {df.loc[df['Temperature'] == temp_max, 'Date'].values[0]}\nminimum temperature: {temp_min}째C -> {df.loc[df['Temperature'] == temp_min, 'Date'].values[0]}')
print(f'-Highest humidity: {humidity_max}% -> {df.loc[df['Humidity'] == humidity_max, 'Date'].values[0]}\nlowest humidity: {humidity_min}% -> {df.loc[df['Humidity'] == humidity_min, 'Date'].values[0]}')
print(f'-Highest wind speed: {wind_max}km/h -> {df.loc[df['Wind Speed'] == wind_max, 'Date'].values[0]}\nlowest wind speed: {wind_min}km/h -> {df.loc[df['Wind Speed'] == wind_min, 'Date'].values[0]}')
```

# Visualize

### i. Line chart

```python
plt.plot(df['Date'],df['Temperature'],marker = 'o' )
plt.title('--Temperature changes over time--')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.gcf().autofmt_xdate()
plt.show
```

### ii. Bar chart

```python
plt.bar(df['Temperature'],df['Humidity'])
plt.title('--Humidity and Temperature--')
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.show()
```

```python
plt.bar(df['Temperature'],df['Wind Speed'],color = 'red')
plt.title('--Temperature and Wind speed--')
plt.xlabel('Temperature')
plt.ylabel('Wind Speed')
plt.show()
```

```python
plt.bar(df['Humidity'],df['Wind Speed'],color = 'green')
plt.title('Humidity and Wind speed')
plt.xlabel('Humdity')
plt.ylabel('Wind Speed')
plt.show()
```

```python
sns.heatmap(cr,annot=True)
```

Overall, the **weather data analysis** provides a clear understanding of how `temperature`, `humidity`, and `wind speed` behave across different days.
The insights suggest that:

-Higher temperatures often come with lower humidity, a common pattern in warm weather.

-Wind speed remains mostly steady and does not strongly depend on temperature or humidity.

-The dataset can be used for further trend analysis or prediction models like forecasting future weather conditions.
