```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('student_marks.csv')
```

```python
data
```

```python
data.isnull().sum()
```

# average marks per subject

```python
subject_avg = data[['Maths','Science','English']].mean()
```

```python
subject_avg
```

# total marks for each student

```python
data['Total'] = data['Maths'] + data['Science'] + data['English']
print(data[['Name', 'Total']])
```

# Identify the topper and the lowest scorer

```python
topper = data.loc[data['Total'].idxmax()]
print('Topper is:', topper['Name'], 'with total marks:',topper['Total'])
```

```python
lowest = data.loc[data['Total'].idxmin()]
print('Lowest scorer is:', lowest['Name'], 'with total marks:',lowest['Total'])
```

# Bar chart

```python
plt.bar(data['Name'],data['Total'])
plt.xlabel('Students')
plt.ylabel('Total Marks')
plt.title('Total Marks for each students')
plt.show()
```

# line chart

```python
plt.plot(subject_avg.index,subject_avg.values, marker = 'o')
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.title("Average Marks Across Subjects")
plt.show()
```

```python
def get_grade(total):
    if total >= 80:
        return 'A'
    elif total >= 60:
        return 'B'
    else:
        return 'C'

data['Grade'] = data['Total'] / 3
data['Grade'] = data['Grade'].apply(get_grade)

print(data[['Name','Total','Grade']])
```
